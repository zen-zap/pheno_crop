### **1. The Wrapper: `.SAFE` Folder**

* **What it is:** The top-level folder. Think of it as the "Product Box."
* **Contains:** Everything—images, metadata, reports.
* **Key File:** `MTD_MSIL2A.xml` (Metadata). This file tells your code exactly where the satellite was, the sun angle, and the cloud cover percentage.

### **2. The Data: `GRANULE` Folder**

* **What it is:** The actual "Tile" of data.
* **Why it's nested:** Sometimes a product covers multiple tiles (though usually just one for L2A).
* **Inside:** You found a folder named `L2A_T16TCK_...`.
* **L2A:** Level 2A (Atmosphere Corrected).
* **T16TCK:** This is the Tile ID for **Champaign, Illinois**. (Just like `T45RUJ` was Bihar).



### **3. The Treasure Chest: `IMG_DATA` Folder**

This is where the actual images live. It is usually split by **resolution** to save space.

* **`R10m/` (10 meters):** **High Res.** Contains **B02, B03, B04, B08**.
* *Usage:* This is 90% of what you need for crop classification.


* **`R20m/` (20 meters):** **Med Res.** Contains **B05, B06, B07** (Red Edge) and **B11, B12** (SWIR).
* *Usage:* "Red Edge" bands are amazing for detecting chlorophyll content, but they are blurrier. SWIR is for water stress.


* **`R60m/` (60 meters):** **Low Res.** Atmospheric bands (B01, B09).
* *Usage:* Useless for you. Ignore these.



### **4. The Quality Check: `QI_DATA` Folder**

* **Inside:** You will see files like `MSK_CLDPRB_20m.jp2` (Cloud Probability Mask).
* **Usage:** Advanced models use this to say "Ignore this pixel, it's a cloud." Your current script does a simpler version of this by ignoring bright white pixels, but this folder is the "official" cloud mask.
