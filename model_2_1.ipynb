### Plot Coordinates Setup Script

```python
import pandas as pd
import ast

gt_path = "/content/drive/MyDrive/pheno_crop/2025_2026_ground_truth.csv"
df = pd.read_csv(gt_path)

unique_plots = df.drop_duplicates(subset=['plot_id'])

plots = []

for _, row in unique_plots.iterrows():
    raw_coords_list = ast.literal_eval(row['raw_coordinates'])[0] 
    
    plots.append({
        "id": row['plot_id'],
        "coords": raw_coords_list
    })

print(f"Successfully extracted coordinates for {len(plots)} unique plots!")

print("\nExample formatted plot:")
print(plots[0])
```
